```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("data/track6_reports_with_entities.csv")
df
```

```python
# Basic Dataset Information
print("Dataset Shape:", df.shape)
print("\n" + "="*50)
print("Column Names and Data Types:")
print("="*50)
print(df.dtypes)
print("\n" + "="*50)
print("First few rows:")
print("="*50)
df.head()
```

```python
# Numeric Columns Distribution
numeric_cols = df.select_dtypes(include=['int64', 'float64', 'int32', 'float32']).columns
print(f"Numeric Columns: {list(numeric_cols)}")
print("="*50)

fig, axes = plt.subplots(nrows=(len(numeric_cols) + 1) // 2, ncols=2, figsize=(14, 5 * ((len(numeric_cols) + 1) // 2)))
axes = axes.flatten()

for idx, col in enumerate(numeric_cols):
    axes[idx].hist(df[col].dropna(), bins=30, edgecolor='black', alpha=0.7)
    axes[idx].set_title(f'Distribution of {col}')
    axes[idx].set_xlabel(col)
    axes[idx].set_ylabel('Frequency')

# Hide extra subplots
for idx in range(len(numeric_cols), len(axes)):
    axes[idx].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
# Duplicate Rows Analysis
print("Duplicate Rows Analysis:")
print("="*50)
duplicates = df.duplicated().sum()
print(f"Total duplicate rows: {duplicates}")
if duplicates > 0:
    print(f"Percentage of duplicates: {(duplicates/len(df))*100:.2f}%")
```

```python
# Categorical Columns Analysis
categorical_cols = df.select_dtypes(include=['object']).columns
print(f"Categorical Columns: {list(categorical_cols)}")
print("="*50)

for col in categorical_cols:
    print(f"\n{col}:")
    print(f"  Unique values: {df[col].nunique()}")
    print(f"  Top 5 values:\n{df[col].value_counts().head()}")
```

```python
# Statistical Summary
print("Statistical Summary:")
print("="*50)
df.describe().T
```

```python
# Missing Values Analysis
import matplotlib.pyplot as plt
import seaborn as sns

missing_data = df.isnull().sum()
missing_percent = (df.isnull().sum() / len(df)) * 100
missing_df = pd.DataFrame({'Missing_Count': missing_data, 'Percentage': missing_percent})
missing_df = missing_df[missing_df['Missing_Count'] > 0].sort_values('Missing_Count', ascending=False)

print("Missing Values:")
print("="*50)
if len(missing_df) > 0:
    print(missing_df)
else:
    print("No missing values found!")
```

```python
# Data Quality Overview
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Missing values percentage
if len(missing_df) > 0:
    missing_plot_data = missing_df['Percentage'].sort_values(ascending=True)
    axes[0].barh(range(len(missing_plot_data)), missing_plot_data.values, color='coral')
    axes[0].set_yticks(range(len(missing_plot_data)))
    axes[0].set_yticklabels(missing_plot_data.index)
    axes[0].set_xlabel('Missing Percentage (%)')
    axes[0].set_title('Missing Values by Column')
else:
    axes[0].text(0.5, 0.5, 'No Missing Values', ha='center', va='center', fontsize=14)
    axes[0].set_title('Missing Values by Column')
    axes[0].axis('off')

# Data type distribution
dtype_counts = df.dtypes.value_counts()
colors = plt.cm.Set3(range(len(dtype_counts)))
axes[1].pie(dtype_counts.values, labels=dtype_counts.index, autopct='%1.1f%%', colors=colors)
axes[1].set_title('Data Type Distribution')

plt.tight_layout()
plt.show()
```

```python
# Box Plots for Numeric Columns
if len(numeric_cols) > 0:
    fig, axes = plt.subplots(nrows=(len(numeric_cols) + 1) // 2, ncols=2, figsize=(14, 5 * ((len(numeric_cols) + 1) // 2)))
    axes = axes.flatten()
    
    for idx, col in enumerate(numeric_cols):
        axes[idx].boxplot(df[col].dropna())
        axes[idx].set_title(f'Box Plot of {col}')
        axes[idx].set_ylabel(col)
    
    for idx in range(len(numeric_cols), len(axes)):
        axes[idx].set_visible(False)
    
    plt.tight_layout()
    plt.show()
else:
    print("No numeric columns found")
```

```python
# Correlation Heatmap
numeric_data = df.select_dtypes(include=['int64', 'float64', 'int32', 'float32'])

if len(numeric_data.columns) > 1:
    plt.figure(figsize=(10, 8))
    correlation = numeric_data.corr()
    sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0, fmt='.2f', 
                square=True, linewidths=1, cbar_kws={"shrink": 0.8})
    plt.title('Correlation Matrix Heatmap')
    plt.tight_layout()
    plt.show()
else:
    print("Not enough numeric columns for correlation matrix")
```
